### Part 1: Data Exploration ###
1. Print the shape and column names of the dataset.

2. Display basic information and summary statistics about the dataset.

3. Show the first 10 rows.

```python
# Part 1: Data Exploration

# 1. print the shape and column names of the dataset
print("Shape:", df.shape)
print("Column Names:", df.columns)

# 2. display basic information and summary statistics
print("Basic info:", df.info())

print("Summary statistics:", df.describe())

# 3. show the first 10 rows
print("First 10 rows:", df.head(10))
```

#### Explanation

- `df.shape` shows the number of rows and columns in the dataset.
- `df.columns` shows all column names in the dataset.
- `df.info()` shows basic information about the dataset such as data types and so on.
- `df.describe()` creates summary statistics for numerical columns like mean, min, max, and so on.
- `df.head(10)` shows the first 10 rows of the dataset.

### Part 2: Indexing and Selection ###
1. Select the names column.

2. Select multiple columns of your choice.

3. Select a single row from the dataset.

4. Select a rectangular section (subset of rows and columns).

```python
# Part 2: Indexing and Selection

# 1. select the names column
print("Names column:", df["Name"])

# 2. select multiple columns of your choice
print("Multiple columns:", df[["Name", "Age", "SibSp"]])

# 3. select a single row from the dataset
print("Single row:", df.loc[0])

# 4. select a rectangular section (subset of rows and columns)
print("Subset of rows and columns:", df.loc[0:4, ["Name", "Age", "SibSp"]])
```

#### Explanation

- `df["Name"]` selects only the Name column from the dataset.
- `df[["Name", "Age", "SibSp"]]` selects multiple columns at the same time.
- `df.loc[0]` selects a single row using its index label.
- `df.loc[0:4, ["Name", "Age", "SibSp"]]` selects a smaller section of the dataset with specific rows and columns.

### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

```python
# Part 3: Filtering

# 1. show only the rows where the age is above 60
print("Passengers older 60:", df[df["Age"] > 60])

# 2. show rows where the fare is above 100 and passenger was in 1st class
print("Passengers fare 100 and 1st class:", df[(df["Fare"] > 100) & (df["Pclass"] == 1)])

# 3. identify and isolate rows with missing values in the age column
print("Rows missing age :", df[df["Age"].isnull()])
```

#### Explanation

- `df[df["Age"] > 60]` filters the dataset and only shows passengers older than 60.
- `&` is used to combine multiple conditions.
- `df[(df["Fare"] > 100) & (df["Pclass"] == 1)]` shows passengers who paid more than 100 and were in first class. as a long condition
- `isnull()` checks for missing values in a column.
- `df[df["Age"].isnull()]` puts alone rows where the age value is missing.

### Part 4: Reindexing and Dropping ###

1. Reverse the row order of the dataset.

2. Remove one column from the dataset.

3. Remove all rows with missing values in the age column.

```python
# Part 4: Reindexing and Dropping

# 1. reverse the row order of the dataset
print("Reversed dataset:", df[::-1])

# 2. remove one column from the dataset
print("Removed Cabin Column:", df.drop(columns=["Cabin"]))

# 3. remove all rows with missing values in the age column
print("rows removed:", df.dropna(subset=["Age"]))
```

#### Explanation

- `df[::-1]` reverses the order of rows.
- `drop(columns=["Cabin"])` removes the Cabin column.
- `dropna(subset=["Age"])` removes rows where the Age column has missing values.
- `subset=["Age"]` means only the Age column is checked for missing values.
