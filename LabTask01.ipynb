### Part 2: Indexing and Selection ###
1. Select the names column.

### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

```python
# Part 3: Filtering

# 1. show only the rows where the age is above 60
print("Passengers older 60:", df[df["Age"] > 60])

# 2. show rows where the fare is above 100 and passenger was in 1st class
print("Passengers fare 100 and 1st class:", df[(df["Fare"] > 100) & (df["Pclass"] == 1)])

# 3. identify and isolate rows with missing values in the age column
print("Rows missing age :", df[df["Age"].isnull()])
```

#### Explanation

- `df[df["Age"] > 60]` filters the dataset and only shows passengers older than 60.
- `&` is used to combine multiple conditions.
- `df[(df["Fare"] > 100) & (df["Pclass"] == 1)]` shows passengers who paid more than 100 and were in first class. as a long condition
- `isnull()` checks for missing values in a column.
- `df[df["Age"].isnull()]` puts alone rows where the age value is missing.

### Part 4: Reindexing and Dropping ###

1. Reverse the row order of the dataset.

2. Remove one column from the dataset.

3. Remove all rows with missing values in the age column.

```python
# Part 4: Reindexing and Dropping

# 1. reverse the row order of the dataset
print("Reversed dataset:", df[::-1])

# 2. remove one column from the dataset
print("Removed Cabin Column:", df.drop(columns=["Cabin"]))

# 3. remove all rows with missing values in the age column
print("rows removed:", df.dropna(subset=["Age"]))
```

#### Explanation

- `df[::-1]` reverses the order of rows.
- `drop(columns=["Cabin"])` removes the Cabin column.
- `dropna(subset=["Age"])` removes rows where the Age column has missing values.
- `subset=["Age"]` means only the Age column is checked for missing values.

### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

```python
import numpy as np
```

```python
# Part 5: Arithmetic and Data Alignment

# 1. create a new column that converts fares from USD to Euros
# using a simple conversion rate of 0.923
df["Fare_Euro"] = df["Fare"] * 0.923

print("Fare converted:", df[["Fare", "Fare_Euro"]])

# 2. subtract the average age from every age value
average_age = df["Age"].mean()
print("Age subtracting average :", df["Age"] - average_age)

# 3. subtract a custom array from two selected columns in dataset
custom_array = np.array([1, 2])
print("\nSubtracting custom array:", df[["SibSp", "Parch"]] - custom_array)
```

#### Explanation

- `df["Fare"] * 0.923` changes fare values from USD to Euros using a conversion rate of 0.923.
- A new column called `Fare_Euro` is created to store the converted values.
- `mean()` calculates the average age of passengers.
- Subtracting the average age from every value compares each passenger age to the overall average.
- `np.array([1, 2])` creates a custom NumPy array.
- The custom array is subtracted from the  columns using data alignment.

### Part 6: Applying Functions ###

1. Create a new column with the number of characters in each name.

2. Calculate the difference between fare and age for each row.

3. Round all numerical values in the dataset to one decimal place.



```python
# Part 6: Applying Functions

# 1. create a new column with the number of characters in each name
df["Name_Length"] = df["Name"].apply(len)
print("Names and character count:", df[["Name", "Name_Length"]])

# 2. calculate the difference between fare and age for each row
df["Fare_Age_Difference"] = df["Fare"] - df["Age"]
print("Difference:", df[["Fare", "Age", "Fare_Age_Difference"]])

# 3. round all numerical values in the dataset to one decimal place
print("rounded values:", df.round(1))
```

#### Explanation

- `apply(len)` applies the `len` function to every name in the Name column.
- A new column called `Name_Length` stores the number of characters in each passenger name.
- `df["Fare"] - df["Age"]` calculates the difference between fare and age for every row.
- `round(1)` rounds all numerical values in the dataset to one decimal place.

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
# Part 7: Mapping and Transformation

# 1. convert categorical values in gender column to numeric codes
df["Sex_Code"] = df["Sex"].map({"male": 0, "female": 1})
print("Gender converted:", df[["Sex", "Sex_Code"]])

# 2. create a new column that categorizes age into groups
def age_group(age):
    if pd.isnull(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

print("Age groups:", df[["Age", "Age_Group"]])
```

#### Explanation

- `map()` converts categorical values into different values using a dictionary as per the cheatsheet.
- Male passengers were converted to `0` and female passengers to `1`.
- A custom function called `age_group()` was created to categorize passengers by age.
- `apply()` applies the custom function to every value in the Age column.
- Missing age values were labeled as `"Unknown"`. to avoid all errors

### Part 8: My Choice ###
1. Part 2 / 1. Select the names column.
2. part 3 / 1. Show only the rows where the age is above 60.
3. part 5 / 1. Create a new column that converts fares from USD to Euros.



```python
# Part 8: Python Comparison

# convert data into a Python list of dictionaries
data_list = df.to_dict("records")


# 1. Part 2 / 1. Select the names column.

# create empty list to store names
names = []

# loop through every row in dataset
for row in data_list:

    # add each name into the list
    names.append(row["Name"])

print(names)



# 2. Part 3 / 1. Show only rows where age is above 60.

# create empty list for passengers older than 60
older_than_60 = []

# loop through dataset
for row in data_list:

    # check that age is not missing and age is above 60
    if row["Age"] == row["Age"] and row["Age"] > 60:

        # add matching rows into new list
        older_than_60.append(row)

print(older_than_60)



# 3. Part 5 / 1. Convert fares from USD to Euros.

# create empty list for converted fares
fare_euro = []

# loop through dataset
for row in data_list:

    # convert fare into Euros using 0.92 conversion rate
    fare_euro.append(row["Fare"] * 0.92)

print(fare_euro)
```

#### Explanation

- `df.to_dict("records")` converts the dataset into a Python list of dictionaries.
- A `for` loop was used to go through each row in the dataset.
- `append()` was used to store values into new lists.
- Pure Python was used instead of Pandas functions and methods.
